```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import plotly.express as px
from PIL import Image
from statistics import mode
import matplotlib.pyplot as plt
```

```python
data = [3, 5, 2, 8, 19, 2, 7, 2, 2, 2]
def st(data):
    print("Mean:", np.mean(data))
    print("Median:", np.median(data))
    if type(data) == list:
        print("Mode:", mode(data))
    else:
        print("Mode:", data.mode()[0]) 
    print("Range:", max(data)-min(data))
    print("Q1:", np.quantile(data, q = 0.25))
    print("Q2:", np.quantile(data, q = 0.5))
    print("Q3:", np.quantile(data, q = 0.75))
    print("IQR:", np.quantile(data, q = 0.75) - np.quantile(data, q = 0.25))
    print("STD:", np.std(data)) 
    print("Coefficient of Variation:", stats.variation(data)) 
    print("Skewness:", stats.skew(data)) 
    print("Kurtosis:", stats.kurtosis(data)) 
st(data)
```

```python
sns.boxplot(data) 
""" 
Any value which is > q3 + 1.5 * (IQR) or < q1 - 1.5 * (IQR) is outlier.
Hence, 19 is outlier.
"""
```

```python
# Item Weight Checking Pattern 
combined[combined["Item_Identifier"] == "FDU37"]["Item_Weight"]
```

## **4. Feature Engineering**

```python
combined.head()
```

```python
# Price per Unit
combined["PPU"] = combined["Item_MRP"] / combined["Item_Weight"]
```

```python
# Perishable / Non-Perishable
alleat = combined["Item_Type"].unique()
```

```python
per = ['Dairy', 'Meat', 'Fruits and Vegetables', 'Breakfast', 'Breads', 'Seafood']
def func(x):
    if x in per:
        return ("Perishables")
    else:
        return ("Non Perishables")
combined["Item_category"] = combined.Item_Type.apply(func)  
```

```python
# Outlet Vintage (No of years the outlet was there)
combined["Outlet_Vintage"] = 2013 - combined["Outlet_Establishment_Year"]
```

```python
# Removing columns that have no use.
combined.drop(["Item_Identifier", "Item_MRP", "Outlet_Establishment_Year"], axis = 1, inplace = True) 
```

```python
# Item Identifiers
combined["Item_Id"] = combined["Item_Identifier"].str[0:2] 
```

```python
combined.head()
```

```python
# DataFrame without Visibilty 0 values.
vis = combined.loc[combined["Item_Visibility"] != 0]
```

```python
vis.shape
```

```python
combined.columns
```

```python
from sklearn.preprocessing import LabelEncoder

cols = ['Item_Fat_Content', 'Outlet_Identifier', "Outlet_Size", "Outlet_Location_Type", "Outlet_Type", "Item_category", "Item_Id"]
 
vis[cols] = vis[cols].apply(LabelEncoder().fit_transform)
```

```python
train = vis.loc[vis["Item_Outlet_Sales"].notnull()]
test = vis.loc[vis["Item_Outlet_Sales"].isnull()]
```

```python
x = train.loc[:, train.columns != "Item_Outlet_Sales"] 
y = train.loc[:, "Item_Outlet_Sales"] 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, random_state = 0, test_size = 0.3)
```

```python
# Feature Scalling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_val = sc.transform(x_val) 
```

```python
x_train
```

```python
# Deep Learning Model
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(500, activation = 'relu', input_dim = 11, kernel_initializer = 'normal'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(1)) 

model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 10, validation_data = (x_val, y_val), verbose = 2)
```
